```python
def prime_factors(n):
    """分解质因数"""
    factors = []
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.append(d)
            n //= d
        d += 1
    if n > 1:
        factors.append(n)
    return factors
```

```python
print(f"{x} = {' × '.join(map(str, prime_factors(x)))}")
```
